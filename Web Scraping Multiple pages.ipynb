```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

df = pd.DataFrame()

base_url = "https://www.scrapethissite.com/pages/forms/?page_num="

page_number = 1

total_pages = 6

while page_number <= total_pages:
    current_page = base_url + str(page_number) + "&per_page=100"
    print(f"Fetching data from {current_page}...")

    
    page = requests.get(current_page)
    print(f"Page status: {page.status_code}")
    if page.status_code != 200:
        print("Error fetching page")
        break

    soup = BeautifulSoup(page.text, 'html.parser')


    table1 = soup.find('table')
    
    if table1:
        print("Table found!")
    else:
        print("Table not found!")
        break


    columns = table1.find_all('th')
    columns_title = [title.text.strip() for title in columns]
    print("Column Names:", columns_title)


    rows = table1.find_all('tr')[1:]
    

    temp_df = pd.DataFrame(columns=columns_title)
    
    for row in rows:
        row_data = row.find_all('td')
        if row_data:
            ind_row_data = [data.text.strip() for data in row_data]
            temp_df.loc[len(temp_df)] = ind_row_data
        else:
            print("No data in this row.")


    df = pd.concat([df, temp_df], ignore_index=True)


    page_number += 1


    time.sleep(2)


df.to_csv('HockeyTeamFinal.csv', index=False)

print("Final DataFrame:")
print(df)
```
